```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import uniform
import math
```

```python
import numpy as np
from scipy.stats import uniform

# --- NOMENCLATURA CONCEPTUAL EN CÓDIGO ---
# La nomenclatura se representa conceptualmente como la definición de la distribución
# con sus parámetros generales (a, b).
# En scipy.stats, 'loc' corresponde a 'a' y 'scale' a 'b - a'.
# Podemos usar una cadena para la representación teórica de la nomenclatura.
nomenclatura_general = "X ~ U(a, b)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Uniforme:")
print("  Función de Densidad de Probabilidad (PDF): f(x) = 1 / (b - a)  [para a <= x <= b, 0 en otro caso]")
print("  Media (μ): (a + b) / 2")
print("  Varianza (σ²): (b - a)² / 12")
print("  Desviación Estándar (σ): sqrt((b - a)² / 12)\n")

# Graficando Uniforme
uniforme = stats.uniform()
x = np.linspace(uniforme.ppf(0.01),
                uniforme.ppf(0.99), 100)
fp = uniforme.pdf(x) # Función de Probabilidad
fig, ax = plt.subplots()
ax.plot(x, fp, '--')
ax.vlines(x, 0, fp, colors='b', lw=5, alpha=0.5)
ax.set_yticks([0., 0.2, 0.4, 0.6, 0.8, 1., 1.2])
plt.title('Distribución Uniforme')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python

nomenclatura_general = "X ~ Tri(a, b, c)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Triangular:")
print("  Función de Densidad de Probabilidad (PDF):")
print("    f(x) = 0                       [para x < a]")
print("    f(x) = 2(x - a) / ((b - a)(c - a)) [para a <= x < c]")
print("    f(x) = 2(b - x) / ((b - a)(b - c)) [para c < x <= b]")
print("    f(x) = 0                       [para x > b]")
print("\n  Media (μ): (a + b + c) / 3")
print("  Varianza (σ²): (a² + b² + c² - ab - ac - bc) / 18")
print("  Desviación Estándar (σ): sqrt((a² + b² + c² - ab - ac - bc) / 18)\n")

# Graficando Distribucion Triangular
df = 50 # parametro de forma.
t = stats.t(df)
x = np.linspace(t.ppf(0.01),
                t.ppf(0.99), 100)
fp = t.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Triangular')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python

nomenclatura_general = "X ~ Exp(λ)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Exponencial:")
print("  Función de Densidad de Probabilidad (PDF): f(x; λ) = λ * e^(-λx)  [para x >= 0, 0 para x < 0]")
print("  Media (μ): 1 / λ")
print("  Varianza (σ²): 1 / λ²")
print("  Desviación Estándar (σ): 1 / λ\n")

# Graficando Exponencial
exponencial = stats.expon()
x = np.linspace(exponencial.ppf(0.01),
                exponencial.ppf(0.99), 100)
fp = exponencial.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Exponencial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python

# La nomenclatura se representa conceptualmente con sus parámetros generales (m, lambda).
nomenclatura_general = "X ~ Erlang(m, λ)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("M-Erlang:")
print("  Parámetros: m (entero positivo), λ (tasa, λ > 0)")
print("  Función de Densidad de Probabilidad (PDF): f(x; m, λ) = (λ^m * x^(m-1) * e^(-λx)) / (m-1)!  [para x >= 0]")
print("  Media (μ): m / λ")
print("  Varianza (σ²): m / λ²")
print("  Desviación Estándar (σ): sqrt(m) / λ")
```

```python

nomenclatura_general = "X ~ Gamma(k, θ)  [o X ~ Gamma(α, β)]"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Gamma:")
print("  Parámetros:")
print("    k (o α): Parámetro de forma (shape parameter), k > 0")
print("    θ (o β): Parámetro de escala (scale parameter), θ > 0")
print("    (Algunas fuentes usan λ = 1/θ como parámetro de tasa)")
print("\n  Función de Densidad de Probabilidad (PDF):")
print("    f(x; k, θ) = (x^(k-1) * e^(-x/θ)) / (Γ(k) * θ^k)  [para x >= 0]")
print("    (Donde Γ(k) es la función Gamma)")
print("\n  Media (μ): k * θ")
print("  Varianza (σ²): k * θ²")
print("  Desviación Estándar (σ): θ * sqrt(k)")

# Graficando Gamma
a = 2.6 # parametro de forma.
gamma = stats.gamma(a)
x = np.linspace(gamma.ppf(0.01),
                gamma.ppf(0.99), 100)
fp = gamma.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Gamma')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Representación de las Expresiones Teóricas (general):")
print("  Parámetros:")
print("    k (o α): Parámetro de forma (shape parameter), k > 0")
print("    λ (o β): Parámetro de escala (scale parameter), λ > 0")
print("\n  Función de Densidad de Probabilidad (PDF):")
print("    f(x; k, λ) = (k/λ) * (x/λ)^(k-1) * e^(-(x/λ)^k)  [para x >= 0]")
print("    (Para x < 0, f(x) = 0)")
print("\n  Media (μ): λ * Γ(1 + 1/k)")
print("    (Donde Γ es la función Gamma)")
print("\n  Varianza (σ²): λ² * [Γ(1 + 2/k) - (Γ(1 + 1/k))²]")
print("\n  Desviación Estándar (σ): λ * sqrt(Γ(1 + 2/k) - (Γ(1 + 1/k))²)")
```

```python
nomenclatura_general = "X ~ N(μ, σ²)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Normal Continua):")
print("  Parámetros:")
print("    μ: Media (mean), μ ∈ ℝ")
print("    σ²: Varianza (variance), σ² > 0")
print("    (σ: Desviación Estándar, σ = sqrt(σ²))")
print("\n  Función de Densidad de Probabilidad (PDF):")
print("    f(x; μ, σ) = (1 / (σ * sqrt(2π))) * e^(-0.5 * ((x - μ) / σ)²)  [para x ∈ ℝ]")
print("\n  Media (μ): μ")
print("  Varianza (σ²): σ²")
print("  Desviación Estándar (σ): σ")

# Graficando Normal
mu, sigma = 0, 0.2 # media y desvio estandar
normal = stats.norm(mu, sigma)
x = np.linspace(normal.ppf(0.01),
                normal.ppf(0.99), 100)
fp = normal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
nomenclatura_general = "X ~ LogN(μ, σ²)  [μ y σ² son los parámetros de la Normal de ln(X)]"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Log-Normal:")
print("  Parámetros:")
print("    μ: Media de la distribución Normal del logaritmo natural de X (μ_lnX), μ ∈ ℝ")
print("    σ²: Varianza de la distribución Normal del logaritmo natural de X (σ²_lnX), σ² > 0")
print("    (σ: Desviación Estándar de ln(X), σ = sqrt(σ²))")
print("\n  Función de Densidad de Probabilidad (PDF):")
print("    f(x; μ, σ) = (1 / (x * σ * sqrt(2π))) * e^(-((ln(x) - μ)² / (2σ²)))  [para x > 0]")
print("    (Para x <= 0, f(x) = 0)")
print("\n  Media de X (μ_X): e^(μ + σ²/2)")
print("  Varianza de X (σ²_X): (e^(σ²) - 1) * e^(2μ + σ²)  o  (e^(σ²) - 1) * (μ_X)²")
print("  Desviación Estándar de X (σ_X): μ_X * sqrt(e^(σ²) - 1)")

# Graficando Log-Normal
sigma = 0.6 # parametro
lognormal = stats.lognorm(sigma)
x = np.linspace(lognormal.ppf(0.01),
                lognormal.ppf(0.99), 100)
fp = lognormal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Log-normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
nomenclatura_general = "X ~ Beta(α, β)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Disribucion Beta:")
print("  Parámetros:")
print("    α: Parámetro de forma 1 (alpha), α > 0")
print("    β: Parámetro de forma 2 (beta), β > 0")
print("\n  Función de Densidad de Probabilidad (PDF):")
print("    f(x; α, β) = (x^(α - 1) * (1 - x)^(β - 1)) / B(α, β)  [para 0 <= x <= 1]")
print("    (Para x < 0 o x > 1, f(x) = 0)")
print("    (Donde B(α, β) es la función Beta, B(α, β) = Γ(α)Γ(β) / Γ(α + β))")
print("\n  Media (μ): α / (α + β)")
print("  Varianza (σ²): (α * β) / ((α + β)² * (α + β + 1))")
print("  Desviación Estándar (σ): sqrt((α * β) / ((α + β)² * (α + β + 1)))")

# Graficando Beta
a, b = 2.3, 0.6 # parametros de forma.
beta = stats.beta(a, b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Beta')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
nomenclatura_general = "X ~ PearsonIII(k, θ, loc)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Pearson Tipo III:")
print("  Parámetros:")
print("    k: Parámetro de forma (shape parameter), k > 0")
print("    θ: Parámetro de escala (scale parameter), θ > 0")
print("    loc: Parámetro de ubicación (location parameter), loc ∈ ℝ")
print("\n  Función de Densidad de Probabilidad (PDF):")
print("    f(x; k, θ, loc) = ((x - loc)^(k-1) * e^(-(x - loc)/θ)) / (Γ(k) * θ^k)  [para x >= loc]")
print("    (Para x < loc, f(x) = 0)")
print("    (Donde Γ(k) es la función Gamma)")
print("\n  Media (μ): k * θ + loc")
print("  Varianza (σ²): k * θ²")
print("  Desviación Estándar (σ): θ * sqrt(k)")
```

```python
nomenclatura_general = "X ~ LogLogistic(α, β)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Log-Logistica:")
print("  Parámetros:")
print("    α: Parámetro de escala (scale parameter), α > 0")
print("    β: Parámetro de forma (shape parameter), β > 0")
print("\n  Función de Densidad de Probabilidad (PDF):")
print("    f(x; α, β) = ( (β/α) * (x/α)^(β-1) ) / (1 + (x/α)^β)²  [para x >= 0]")
print("    (Para x < 0, f(x) = 0)")
print("\n  Media (μ): (Existe si β > 1)")
print("    μ = (α * π) / (β * sin(π/β))")
print("\n  Varianza (σ²): (Existe si β > 2)")
print("    σ² = α² * [ (π * 2/β) / sin(2π/β)  -  ( (π/β) / sin(π/β) )² ]")
print("\n  Desviación Estándar (σ): sqrt(σ²)")
```

```python
nomenclatura_general = "X ~ Johnson(γ, δ, ξ, λ)  [Tipo: SB, SU, o SL]"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Johnson:")
print("  Parámetros:")
print("    γ (gamma): Parámetro de forma (skewness)")
print("    δ (delta): Parámetro de forma (kurtosis/scale)")
print("    ξ (xi): Parámetro de ubicación (location)")
print("    λ (lambda): Parámetro de escala (scale)")
print("\n  Transformación General: Z = f(X; γ, δ, ξ, λ), donde Z ~ N(0, 1)")
print("\n  Tipos de Johnson y sus transformaciones:")
print("    1. Johnson SB (Bounded):")
print("       Rango: ξ < X < ξ + λ")
print("       Z = γ + δ * ln( (X - ξ) / (ξ + λ - X) )")
print("\n    2. Johnson SU (Unbounded):")
print("       Rango: -∞ < X < ∞")
print("       Z = γ + δ * arsinh( (X - ξ) / λ )")
print("\n    3. Johnson SL (Log-Normal):")
print("       Rango: X > ξ")
print("       Z = γ + δ * ln( (X - ξ) / λ )")

print("\n  Media (μ): No hay una fórmula algebraica simple y general para todos los tipos.")
print("             Se calcula numéricamente o se deriva de propiedades de la transformación inversa.")
print("             (Para Johnson SL, es similar a la Log-Normal: μ_X = ξ + λ * e^(-γ/δ + 1/(2δ²)) )")
print("\n  Varianza (σ²): No hay una fórmula algebraica simple y general para todos los tipos.")
print("               Se calcula numéricamente.")
print("               (Para Johnson SL, es similar a la Log-Normal: σ²_X = λ² * (e^(1/δ²) - 1) * e^(-2γ/δ + 1/δ²) )")
print("\n  Desviación Estándar (σ): sqrt(Varianza).")
print("\nNota: Debido a la complejidad de las transformaciones, los valores de la media, varianza y desviación estándar")
print("de una distribución de Johnson suelen obtenerse a través de software estadístico que realiza los cálculos necesarios internamente.")
```

```python
nomenclatura_general = "X ~ U_D(a, b)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Uniforme Discreta:")
print("  Parámetros:")
print("    a: Valor mínimo (entero)")
print("    b: Valor máximo (entero), b >= a")
print("    N: Número total de resultados posibles = b - a + 1")
print("\n  Función de Masa de Probabilidad (PMF):")
print("    P(X=x) = 1 / (b - a + 1)  o  1 / N  [para x en {a, a+1, ..., b}]")
print("    (Para otros valores de x, P(X=x) = 0)")
print("\n  Media (μ): (a + b) / 2")
print("  Varianza (σ²): ( (b - a + 1)² - 1 ) / 12  o  (N² - 1) / 12")
print("  Desviación Estándar (σ): sqrt( ( (b - a + 1)² - 1 ) / 12 )  o  sqrt( (N² - 1) / 12 )")

```

```python
nomenclatura_general = "X ~ Ber(p)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Bernoulli:")
print("  Parámetros:")
print("    p: Probabilidad de éxito, p ∈ [0, 1]")
print("\n  Valores Posibles de X: {0, 1}")
print("\n  Función de Masa de Probabilidad (PMF):")
print("    P(X=x) = p  [si x = 1]")
print("    P(X=x) = 1 - p  [si x = 0]")
print("    P(X=x) = 0  [en otro caso]")
print("    O de forma compacta: P(X=x) = p^x * (1-p)^(1-x)  [para x en {0, 1}]")
print("\n  Media (μ): p")
print("  Varianza (σ²): p * (1 - p)")
print("  Desviación Estándar (σ): sqrt(p * (1 - p))")

# Graficando Bernoulli
p =  0.5 # parametro de forma
bernoulli = stats.bernoulli(p)
x = np.arange(-1, 3)
fmp = bernoulli.pmf(x) # Función de Masa de Probabilidad
fig, ax = plt.subplots()
ax.plot(x, fmp, 'bo')
ax.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
ax.set_yticks([0., 0.2, 0.4, 0.6])
plt.title('Distribución Bernoulli')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
nomenclatura_general = "X ~ Geo(p)  [X = número de ensayos hasta el primer éxito]"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Geometrica:")
print("  Parámetros:")
print("    p: Probabilidad de éxito en un solo ensayo, p ∈ (0, 1]")
print("\n  Valores Posibles de X: {1, 2, 3, ...}")
print("\n  Función de Masa de Probabilidad (PMF):")
print("    P(X=x) = (1 - p)^(x-1) * p  [para x = 1, 2, 3, ...]")
print("\n  Media (μ): 1 / p")
print("  Varianza (σ²): (1 - p) / p²")
print("  Desviación Estándar (σ): sqrt((1 - p) / p²)  o  sqrt(1 - p) / p")

# Graficando Geométrica
p =  0.3 # parametro de forma
geometrica = stats.geom(p) # Distribución
x = np.arange(geometrica.ppf(0.01),
              geometrica.ppf(0.99))
fmp = geometrica.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Geométrica')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
nomenclatura_general = "X ~ NB(r, p)  [X = número de fallos antes del r-ésimo éxito]"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Distribucion Binomial Negativa:")
print("  Parámetros:")
print("    r: Número de éxitos deseados (entero positivo), r ∈ {1, 2, 3, ...}")
print("    p: Probabilidad de éxito en un solo ensayo, p ∈ (0, 1]")
print("\n  Valores Posibles de X: {0, 1, 2, 3, ...} (número de fallos)")
print("\n  Función de Masa de Probabilidad (PMF):")
print("    P(X=x) = C(x + r - 1, x) * p^r * (1 - p)^x  [para x = 0, 1, 2, ...]")
print("    (Donde C(n, k) es el coeficiente binomial n choose k)")
print("\n  Media (μ): (r * (1 - p)) / p")
print("  Varianza (σ²): (r * (1 - p)) / p²")
print("  Desviación Estándar (σ): sqrt((r * (1 - p)) / p²)  o  sqrt(r * (1 - p)) / p")
```

```python
nomenclatura_general = "X ~ Poi(λ)"
print(f"Nomenclatura General: {nomenclatura_general}\n")

print("Representación de las Expresiones Teóricas (general):")
print("  Parámetros:")
print("    λ: Tasa promedio de eventos (rate parameter), λ > 0")
print("\n  Valores Posibles de X: {0, 1, 2, 3, ...}")
print("\n  Función de Masa de Probabilidad (PMF):")
print("    P(X=x) = (e^(-λ) * λ^x) / x!  [para x = 0, 1, 2, ...]")
print("\n  Media (μ): λ")
print("  Varianza (σ²): λ")
print("  Desviación Estándar (σ): sqrt(λ)")

# Graficando Poisson
mu =  3.6 # parametro de forma
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
