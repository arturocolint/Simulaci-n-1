```python
# Generador Congruencial Mixto
import pandas as pd
from tabulate import tabulate # libreria tabulate para mejorar la tabla

def generador(a, c, m, x0):
    x = ((a * x0 )+ c) % m
    Vn = x / m
    return x, Vn

# Valores Iniciales
a = 1
c = 121
m = 30
x0 = 17

# Iniciamos el ciclo para encontrar el periodo del generador
x = x0
tabla = []
valores_vistos = {}  # Diciconario eficiente para encontrar el valor repetido sin recorrer cada iteraciòn


for i in range(m + 1):  # Aumentar el rango para poder localizar el ciclo de vida
    x, Vn = generador(a, c, m, x)

    if x in valores_vistos:  #Valor para detectar el ciclo
        ciclo_inicio = valores_vistos[x] #Ingerar cada valor generado
        ciclo_longitud = i - ciclo_inicio # Calculo de la longitud del ciclo.
        print(f"Se detectó un ciclo en la iteración {ciclo_inicio + 1} con longitud {ciclo_longitud}")
        break

    valores_vistos[x] = i # Lugar en que esta x por primera vez
    tabla.append([i + 1, x, Vn])

# Incluimos el DataFrame para organizar y ordenar los datos
df = pd.DataFrame(tabla, columns=['Iteración', 'x', 'Vn'])

# Imprimir tabla
print(tabulate(df, headers='keys', tablefmt='grid'))
```

```python
from tabulate import tabulate

def cuadrado_medio(x0, cantidad):

    numeros_aleatorios = []
    num = x0
    ciclo_vida = []

    for i in range(1, cantidad + 1):
        cuadrado = num ** 2 #elevamos nuestra semilla al cuadrado

        # Convierte el cuadrado a una cadena y extrae la parte central
        cuadrado_cad = str(cuadrado).zfill(8)
        mitad = len(cuadrado_cad) // 2
        num = int(cuadrado_cad[mitad-2:mitad+2])  # Toma 4 dígitos del centro

        num_normalizado = num / 10000  # Asumimos una normalización entre 0 y 1

        # Agrega el número generado, su versión normalizada y la iteración
        ciclo_vida.append([i, num, num_normalizado])

    return ciclo_vida, numeros_aleatorios

# Función para imprimir el ciclo de vida de la semilla y los números generados
def imprimir_ciclo_vida(ciclo_vida):
    # Imprimir la tabla usando tabulate
    headers = ["Iteración", "Número", "Número Normalizado"]
    print(tabulate(ciclo_vida, headers=headers, tablefmt="grid"))



# Variables
x0 = 1998
cantidad = 100
ciclo_vida, resultados = cuadrado_medio(x0, cantidad)

# Imprimir la tabla con los resultados
print("Ciclo de vida de la semilla:")
imprimir_ciclo_vida(ciclo_vida)
```
