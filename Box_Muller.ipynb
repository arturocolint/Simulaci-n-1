```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
#Numeros a generar
```

```python
U1 = np.random.rand(n)
U2 = np.random.rand(n)
#Generacion de vectores de numeros aleatorioes entre 0 y 1
```

```python
Z0 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)
Z1 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * np.pi * U2)
#Forma en codigo de las escuaciones de Box Muller
```

```python
plt.hist(Z0, bins=50, alpha=0.6, label='Z0')
plt.hist(Z1, bins=50, alpha=0.6, label='Z1')
plt.title('Distribuci√≥n normal generada con Box-Muller')
plt.legend()
plt.grid(True)
plt.show()
#Histograma
```
